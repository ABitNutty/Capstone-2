```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
import pickle
from wordcloud import WordCloud
```

```python
# Load in dictionaries of word importance
importance = pickle.load( open( "log_reg_influential_words.pickle", "rb" ) )
```

```python
importance
```

```python
# Create the wordcloud object
goodcloud = WordCloud(width=2048, height=2048, margin=0, background_color="white").generate_from_frequencies(importance['Good'])

badcloud = WordCloud(width=2048, height=2048, margin=0, background_color="white").generate_from_frequencies(importance['Bad'])
```

```python
# Display the generated image:
plt.imshow(goodcloud, interpolation='bilinear')
plt.axis("off")
plt.title('Good Reviews')
plt.show()
```

```python
# Display the generated image:
plt.imshow(badcloud, interpolation='bilinear')
plt.axis("off")
plt.title('Bad Reviews')
plt.show()
```
